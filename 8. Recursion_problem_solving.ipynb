Task 1: 

A. Write a method/function called task1A() that uses a while loop to print from 1 to 10.

B. Convert the previous task into a recursive function called task1B_recursive(). 

C. Write a method/function called task1C() that uses a while loop to print from 1 to N, here N is a user input. 

D. Convert the previous task into a recursive function called task1D_recursive().

```python
def task1A():
    i = 1
    while i < 11:
        print(i)
        i+=1

def task1B_recursive(n):
    if n == 11:
        return ""
    return f"{n}\n{task1B_recursive(n+1)}"

def task1C(n):
    i = 1
    while i < n+1:
        print(i)
        i+=1

def task1D_recursive(n):
    if n == 0:
        return
    task1D_recursive(n-1)
    print(n)
```

```python
task1A()
print(task1B_recursive(1))
task1C(int(input()))
task1D_recursive(int(input()))
```

```python
def task2A( arr ):
    i = 0
    while i < len(arr):
        print(arr[i], end=" ")
        i+=1

def task2B_recursive( arr,i ):
    if i == len(arr):
        return
    print(arr[i], end=" ")
    task2B_recursive(arr,i+1)

def task2C( arr ):
    sum = 0
    for i in arr:
        sum+=i
    return sum

def task2D_recursive( arr ):
    def helper(index,sum):
        if index == len(arr):
            return sum
        return helper(index+1,sum+arr[index])
    return arr[0]+helper(1,0)

def task2E( arr ):
    m_odd = 1
    count = 0
    s_even = 0
    for i in arr:
        if i%2 == 0:
            s_even+=i
        else:
            m_odd*=i
            count+=1
    if count == 0:
        return -s_even
    return m_odd - s_even

def task2F_recursive( arr ): #with 2 helper function, ignore if all the elements are even

    def helper_odd(arr,i,prod):
        if i == len(arr)-1:
            if arr[i]%2 != 0:
                return prod*arr[i]
            else:
                return prod
        if arr[i]%2 != 0:
            return helper_odd(arr,i+1,prod*arr[i])
        else:
            return helper_odd(arr,i+1,prod)

    def helper_even(arr,i,sum):
        if i == len(arr)-1:
            if arr[i]%2 == 0:
                return sum+arr[i]
            else:
                return sum
        if arr[i]%2 == 0:
            return helper_even(arr,i+1,sum+arr[i])
        else:
            return helper_even(arr,i+1,sum)
        
    return helper_odd(arr,0,1) - helper_even(arr,0,0)

def task2F_recursive_1helper(arr): #ignores the case where all the elements are even
    def helper(sum, prod, i):
        if i == len(arr)-1:
            if arr[i]%2 == 0:
                sum+=arr[i]
                return prod - sum
            else:
                prod*=arr[i]
                return prod-sum
        if arr[i]%2 == 0:
            return helper(sum+arr[i], prod, i+1)
        else:
            return helper(sum, prod*arr[i], i+1)
    return helper(0,1,0)
```
